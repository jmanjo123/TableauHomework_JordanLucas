```python
import pandas as pd
```

```python
# Read each dataset using Pandas
file1 = "Resources/JC-201812-citibike-tripdata.csv"
file2 = "Resources/JC-201811-citibike-tripdata.csv"
file3 = "Resources/JC-201810-citibike-tripdata.csv"

df_dec = pd.read_csv(file1)
df_nov = pd.read_csv(file2)
df_oct = pd.read_csv(file3)
```

```python
# Merge the October and November datasets
merged_df1 = df_oct.append(df_nov)
merged_df1.head()
```

```python
len(merged_df1.index)
```

```python
# Merge the existing merged dataset with the December data to make the final set of data
df = merged_df1.append(df_dec)
df.head()
```

```python
len(df.index)
```

```python
# 0, 1, and 2 are unhelpful indicators for gender so I changed the values to reflect the actual gender (as detailed in the documentation).
df.loc[df.gender == 0, 'gender'] = "Unknown"
df.loc[df.gender == 1, 'gender'] = "Male"
df.loc[df.gender == 2, 'gender'] = "Female"

df.head()
```

```python
# I figured that having columns containing tripduration time conversions could help to make things more readable in a visualization.  So I added those to the dataset.
df["tripduration(minutes)"] = df['tripduration']/60
df["tripduration(hours)"] = df['tripduration']/3600

df.head()
```

```python
# I added a record id field to uniquely identify each record in the dataset.  I then reordered the dataset to be easier to read.
df["record_id"] = range(0, len(df.index))

df = df[['record_id', 'starttime', 'stoptime', 'tripduration', 'tripduration(minutes)', 
         'tripduration(hours)', 'bikeid', 'start station id', 'start station name', 
         'start station latitude', 'start station longitude', 'end station id', 'end station name', 
         'end station latitude', 'end station longitude', 'usertype', 'birth year', 'gender']]

df.head()
```

```python
# Take the dataset and save it to a new csv file.
df.to_csv("Resources/CitiBike-2018-Q4-TripData.csv", index=False, header=True)
```
